axis=0: Refers to rows (index labels).
axis=1: Refers to columns (column labels).

```python
import pandas as pd
df=pd.read_csv('dataset.csv')
df=df.drop(['car_ID'],axis=1)

df.head()
```

```python
xdata=df.drop(['price'],axis=1)
numericalCol=xdata.select_dtypes(exclude=['object']).columns
X=xdata[numericalCol]
y=df['price']
type(X)
```

```python
from sklearn.preprocessing import scale
cols=X.columns
X=pd.DataFrame(scale(X))
X.columns=cols
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
modelR=RandomForestRegressor()
modelR.fit(X_train,y_train)
y_pred=modelR.predict(X_test)

from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred)*100)
```
